```python
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ticker = 'MSFT'

ticker_obj = yf.Ticker(ticker)

data = ticker_obj.history(start='2021-01-01', end='2021-12-31')

data.head()
```

```python
data['delta'] = (data['Close'] - data['Open'] )/data['Open']
data['mean'] = (data['Close'] + data['Open'] )/2
data.reset_index(inplace=True, drop=False)
data.head()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.set(style="white")

ax.fill_between(data['Date'], data['Low'], data['High'], alpha=0.2)
ax.plot(data['Date'], data['mean'],)


plt.show()
```
